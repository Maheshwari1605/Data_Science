```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
movies=pd.read_csv("../Association rules/my_movies.csv")
movies.head()
```

```python
movies.shape
```

```python
records = []
for i in range(0, 10):
    records.append([str(movies.values[i,j]) for j in range(0, 14)])
```

```python
association_rules = apriori(records, min_support=0.005)
association_results = list(association_rules)
```
